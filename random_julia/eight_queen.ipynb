```julia
function columnfree(board:: AbstractArray{Bool}, column:: Int)
    for i in 1:board.dims[1]
        if board[column, i]
            return false 
        end
    end
    return true
end
```

```julia
function diagonalsfree(board:: AbstractArray{Bool}, cell:: Tuple)
    return (
        topleftfree(board, cell)
        && bottomleftfree(board, cell)
        && toprightfree(board, cell)
        && bottomrightfree(board, cell)
    )
end
```

```julia
function checkdiagonal(board:: AbstractArray{Bool}, cell:: Tuple, offsetx, offsety)
    x = cell[1] + offsetx
    y = cell[2] + offsety
    board_limit = board.dims[1]+1
    while (0 < x < board_limit && 0 < y < board_limit)
        if board[x,y]
            return false
        end
        x += offsetx
        y += offsety
    end
    return true
end
```

```julia
function topleftfree(board, cell)
    checkdiagonal(board, cell, -1, -1)
end
```

```julia
function bottomrightfree(board, cell)
    checkdiagonal(board, cell, 1, 1)
end
```

```julia
function toprightfree(board, cell)
    checkdiagonal(board, cell, -1, 1)
end
```

```julia
function bottomleftfree(board, cell)
    checkdiagonal(board, cell, 1, -1)
end
```

```julia
function cellsafe(board :: AbstractArray{Bool}, cell :: Tuple)
    return (
        diagonalsfree(board, cell)
        && rowfree(board, cell)
    )
end
```

```julia
function findRestQueens!(board, column = 1)
    if column == board.dims[1] + 1
        global solutions
        solutions = solutions +1
        println("Solution", solutions)
        println(board)
        println()
        return true
    end
    if columnfree(board, column)
        for row in 1:board.dims[1]
            if(cellsafe(board, (row, column)))
                guessboard = copy(board)
                guessboard[row, column] = true
                findRestQueens!(guessboard, column + 1)
            end
        end
        return false
    else 
        findRestQueens!(board, column +1)
    end
end
```
