```python
import timeit
```

```python
time_mult = []

product = '5*5'

for x in (range(150)):
    t = timeit.timeit(f'{product}', number=10000, setup = 'import numpy as np')
    time_mult.append(t)
    product += '*5'
```

```python
time_power = []

for i in (range(2, 152)):
    operation = f'5**{i}'
    t = timeit.timeit(f'{operation}', number = 10000, setup = 'import numpy as np')
    time_power.append(t)
```

```python
import numpy as np
```

```python
time_numpy = []

for i in (range(2, 152)):
    operation = f'power(5,{i})'
    t = timeit.timeit(f'{operation}', number = 10000, setup='from numpy import power')
    time_numpy.append(t)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
plt.figure(figsize = (20,20))
plt.plot((time_mult), c = 'red', label = 'multiplication')
plt.plot((time_power), c = 'green', label = 'python power')
plt.plot((time_numpy), c = 'blue', label = 'numpy')
plt.legend()

plt.xlabel('5 ^ x')
plt.ylabel('time in seconds')

plt.savefig('power_comparison.png')
plt.show()
```
