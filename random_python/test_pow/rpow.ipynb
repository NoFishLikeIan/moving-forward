```r
execute_prod <- function(i) {
    start <- Sys.time()
    5*5*5*5
    end <- Sys.time()
    
    return(end-start)
}
```

```r
v = vector()
for (i in 1:20000) {
    v[i] <- execute_prod()
}
```

```r
execute_pow <- function(i) {
    start <- Sys.time()
    5**4
    end <- Sys.time()
    
    return(end-start)
}
```

```r
v_pow = vector()
for (i in 1:20000) {
    v_pow[i] <- execute_pow()
}
```

```r
x <- 1:length(v_pow)
y <- v_pow/v
```

```r
data = data.frame(x,y)
```

```r
library(ggplot2)

ggplot(data ,aes(x=x, y=y))+
    geom_point(shape=1)
```
